```python
import pandas as pd

file_path = "../data/raw/1. RIE OC1.xlsx"
rie = pd.read_excel(file_path, sheet_name="1. RIE", header=9)
rie.head()
```

```python
rie = rie.dropna(axis=1, how="all")
rie = rie.dropna(subset=["FECHA"]) 

rie["FECHA"] = pd.to_datetime(rie["FECHA"], errors="coerce")

num_cols = ["HRS. ESPERA", "HRS. TALLER", "HRS TALLER AFECTA", "PARADA"]
for c in num_cols:
    if c in rie.columns:
        rie[c] = pd.to_numeric(rie[c], errors="coerce").fillna(0)
```

```python
# Para horometros
raw_horo = pd.read_excel(file_path, sheet_name="2. HOROMETROS", header=None)
raw_horo.head(20)
```

```python
horo = pd.read_excel(file_path, sheet_name="2. HOROMETROS", header=9)
horo.head()
```

```python
horo = horo.loc[:, ~horo.columns.astype(str).str.contains("^Unnamed")]
```

```python
horo["FECHA"] = pd.to_datetime(horo["FECHA"], errors="coerce")
cols_num = [
    "HORAS PROGRAMADAS",
    "TOTAL H. DIESEL",
    "TOTAL MAX ELECTRICO"
]
for c in cols_num:
    if c in horo.columns:
        horo[c] = pd.to_numeric(horo[c], errors="coerce").fillna(0)

horo["HRS_TRAB"] = horo["TOTAL H. DIESEL"] + horo["TOTAL MAX ELECTRICO"]

horo["UT"] = np.where(
    horo["HORAS PROGRAMADAS"] > 0,
    horo["HRS_TRAB"] / horo["HORAS PROGRAMADAS"],
    0
)
```

```python
horo.head()
```

```python
#KPI RIE
import numpy as np
#DM
rie["DM"] = (10 - rie["HRS TALLER AFECTA"]) / 10
#MTTR
rie["MTTR"] = np.where(
    rie["PROG"].astype(str).str.upper().str.strip() == "P",
    0,
    np.where(rie["PARADA"] > 0, rie["HRS TALLER AFECTA"] / rie["PARADA"], 0)
)
```

```python
rie_agg = (
    rie.groupby(["FECHA", "TURNO", "COD EQ INC", "COD EQ MINA"], as_index=False)
    .agg(
        HRS_TALLER_AFECTA=("HRS TALLER AFECTA", "sum"),
        PARADA=("PARADA", "sum"),
        DM=("DM", "mean"),
        MTTR=("MTTR", "mean"),
        N_INTERV=("CODIGO OT", "count") if "CODIGO OT" in rie.columns else ("PARADA", "size")
    )
)

rie_agg.head()
```

```python
# UNE HORO + RIE
base = horo.merge(
    rie_agg,
    on=["FECHA", "TURNO", "COD EQ INC", "COD EQ MINA"],
    how="left"
)

# Si no hubo intervención, rellena con 0
for c in ["HRS_TALLER_AFECTA", "PARADA", "DM", "MTTR", "N_INTERV"]:
    if c in base.columns:
        base[c] = base[c].fillna(0)

base.head()
```

```python
#MTBF Y R(T)
base["MTBF"] = np.where(
    base["PARADA"] > 0,
    base["HRS_TRAB"] / base["PARADA"],
    base["HRS_TRAB"]
)

base["R_t"] = np.exp((-10 * base["UT"]) / base["MTBF"].replace(0, np.nan))
base["R_t"] = base["R_t"].fillna(0)
```

```python
#DM
def semaforo_dm(x):
    if x <= 0.80: return "MALO 🔴"
    if x <= 0.85: return "MEDIO 🟡"
    return "BUENO 🟢"

base["DM_STATUS"] = base["DM"].apply(semaforo_dm)
```

```python
#R(T)
def semaforo_rt(x):
    if x <= 0.50: return "MALO 🔴"
    if x <= 0.71: return "MEDIO 🟡"
    return "BUENO 🟢"

base["RT_STATUS"] = base["R_t"].apply(semaforo_rt)
```

```python
base.to_csv("../data/procesados/base_kpis_turno.csv", index=False)
```

```python
kpis_equipo = (
    base.groupby("COD EQ INC", as_index=False)
    .agg(
        DM=("DM", "mean"),
        UT=("UT", "mean"),
        MTBF=("MTBF", "mean"),
        MTTR=("MTTR", "mean"),
        R_t=("R_t", "mean"),
        PARADAS=("PARADA", "sum"),
        HRS_TRAB=("HRS_TRAB", "sum"),
        INTERV=("N_INTERV", "sum")
    )
)

kpis_equipo.to_csv("../data/procesados/kpis_por_equipo.csv", index=False)
kpis_equipo.sort_values("DM").head(10)
```

```python
#ALERTAS AUTOMATICAS
# DM BAJO <0.85
umbral_dm = 0.85
alert_dm = (
    kpis_equipo[kpis_equipo["DM"] < umbral_dm]
    .sort_values("DM")
)
alert_dm.head(20)
```

```python
#R(T)<0.50
umbral_rt = 0.50
alert_rt = (
    kpis_equipo[kpis_equipo["R_t"] < umbral_rt]
    .sort_values("R_t")
)
alert_rt.head(20)
```

```python
#MUCHAS PARADAS
umbral_paradas = 100
alert_paradas = (
    kpis_equipo[kpis_equipo["PARADAS"] >= umbral_paradas]
    .sort_values("PARADAS", ascending=False)
)
alert_paradas.head(20)
```

```python
#NO EJECUCUION DE MTTOS PROGRAMADOS
rie["ESTADO"] = rie["ESTADO"].astype(str).str.upper().str.strip()
rie["PROG"] = rie["PROG"].astype(str).str.upper().str.strip()
no_ejecutados = rie[(rie["PROG"] == "P") & (rie["ESTADO"] != "INTERVENIDO")]
alert_no_ej = (
    no_ejecutados.groupby("COD EQ INC", as_index=False)
    .size()
    .rename(columns={"size": "P_NO_EJECUTADOS"})
    .sort_values("P_NO_EJECUTADOS", ascending=False)
)
alert_no_ej.head(20)
```

```python
alert_dm.to_csv("../data/procesados/alerta_dm_bajo.csv", index=False)
alert_rt.to_csv("../data/procesados/alerta_confiabilidad_baja.csv", index=False)
alert_paradas.to_csv("../data/procesados/alerta_paradas_altas.csv", index=False)
alert_no_ej.to_csv("../data/procesados/alerta_preventivos_no_ejecutados.csv", index=False)
```
